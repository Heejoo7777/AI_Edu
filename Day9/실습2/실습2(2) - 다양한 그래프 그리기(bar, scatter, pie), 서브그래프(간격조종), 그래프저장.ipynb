```python
#막대 그래프 (bar)
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

x = ['Nana', 'Bo', 'Bibi']
y = [30, 80, 100]

plt.title('class1 수학성적')
plt.bar(x,y) #막대 그래프
plt.ylabel('성적')
plt.show()
```

```python
#산포 그래프 (scatter)
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

x = ['1월', '2월', '3월', '4월', '5월', '6월', '7월','8월','9월','10월','11월','12월' ]
y = [30, 80, 100, 70, 90, 97, 58, 60, 54, 10, 25, 60] 
plt.title('연간 월별 평균 미세먼지')
plt.scatter(x, y) #산포 그래프
plt.ylabel('미세먼지')
plt.show()
```

```python
#파이 그래프 (pie)
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

food = ['한식', '중식', '일식', '양식', '기타']
portion = [30, 20, 15, 25, 10]

#explode=(0.1, 0, 0.1, 0, 0) 로 한식과 중식이 0.1만큼 튀어나오게 만듦
#autopct='%1.1f%%'는 소숫점 이하 1자리수의 부동소수점으로 표시(1.1의 1은 소숫점 이상 자리가 1자리 이상되게, %%는 %출력위해)
plt.pie(portion, explode=(0.1, 0, 0.1, 0, 0), labels=food, autopct='%1.1f%%', shadow=True, startangle=90) 
plt.title('한국인들의 음식 취향')

plt.show()
```

```python
#서브 그래프 그리기 (subplots)
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

x = ['1월', '2월', '3월', '4월', '5월', '6월', '7월','8월','9월','10월','11월','12월' ]
y1 = [30, 80, 100, 70, 90, 97, 58, 60, 54, 10, 25, 60] 
y2 = [37, 72, 57, 43, 98, 67, 48, 20, 15, 70, 65, 55] 

#nrows=1, ncols=2 는 1행 2열, 2개의 그래프 의미
#figsize=(9,3)는 가로 9인치, 세로 3인치
#sharex=True, sharey=True 는 두 서브 그래프의 X와 Y축의 데이터 값이 동일
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9,3), sharex=True, sharey=True) #근데 왜 fig와 axs 두군데에다 저장?
                                                                                     #fig는 전체 그래프, axs는 서브그래프 가리킬 용도
                                                                                     #fig = fig.figure(figure=(9,3))
                                                                                     #axs = plt.subplot(sharex=True,sharey=True)

#첫번째 서브그래프 axs[0]
ax = axs[0]
ax.scatter(x,y1)
ax.set_title('2020')

#두번째 서브그래프 axs[1]
ax = axs[1]
ax.scatter(x,y2)
ax.set_title('2021')

fig.suptitle('연간 월별 평균 미세먼지') #전체 그래프의 제목
plt.show()
```

```python
#그래프 사이 간격 조종하기
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

x = list(range(1,6)) #참고: x와 y의 갯수가 같아야함!!!
y = list(range(10, 80, 15))

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(9,5), sharex=True, sharey=True)

#첫번째 서브그래프: 선그래프 plot
ax = axs[0][0]
ax.plot(x,y)
ax.set_title('선그래프1')

#두번째 서브그래프: 선그래프 + 선스타일 바꿈
ax = axs[0][1]
ax.plot(x,y, color='red', linestyle='--', marker='o')
ax.set_title('선그래프2')

#세번째 서브그래프: 막대그래프 bar
ax = axs[1][0]
ax.bar(x,y)
ax.set_title('막대그래프')

#네번째 서브그래프: 산포그래프 scatter
ax = axs[1][1]
ax.scatter(x,y)
ax.set_title('산포그래프')

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.2, hspace=0.3) #좋은 간격, 그냥 이대로 써
plt.savefig('4개 그래프.png', dpi=300) #그래프 이미지 저장 + 해상도 설정(고해상도 이미지, 300dpi는 1 inch^2에 300개 점 사용)
plt.show()
```
