```python
#간단 그래프그리기 
import matplotlib.pyplot as plt #Matplotlib패키지의 Pyplot모듈을 plt로 임포트
x = [1, 2, 3]
y = [25, 35, 65]
plt.plot(x, y)
plt.show()
```

```python
#그래프 제목(title) + x축&y축 레이블 (xlabel, ylabel)
import matplotlib.pyplot as plt #Matplotlib패키지의 Pyplot모듈을 plt로 임포트
x = ['Nana', 'Bo', 'Bibi']
y = [65, 80, 100]
plt.plot(x, y) #그래프
plt.title('Math Score of class1')#상단 중앙에 그래프의 제목 표시
plt.xlabel('Student Name') #x축 레이블
plt.ylabel('Score') #y축 레이블
plt.show()
```

```python
#한글 폰트 사용하기(from matplotlib import rc, rc함수)
import matplotlib.pyplot as plt #Matplotlib패키지의 Pyplot모듈을 plt로 임포트
from matplotlib import rc #rc모듈 임포트
rc('font', family='Malgun Gothic') #rc함수를 이용해 font를 맑은 고딕으로
x = ['Nana', 'Bo', 'Bibi']
y = [65, 80, 100]
plt.plot(x, y) #그래프
plt.title('class1의 수학성적')#상단 중앙에 그래프의 제목 표시
plt.xlabel('학생 이름') #x축 레이블
plt.ylabel('성적') #y축 레이블
plt.show()
```

```python
#볌례 표시하기 (plt.legend()의 loc이용)
import matplotlib.pyplot as plt #Matplotlib패키지의 Pyplot모듈을 plt로 임포트
from matplotlib import rc #rc모듈 임포트
rc('font', family='Malgun Gothic') #rc함수를 이용해 font를 맑은 고딕으로
x = ['Nana', 'Bo', 'Bibi']
y1 = [65, 80, 100]
y2 = [70, 50, 85]
plt.plot(x, y1, label='국어') #그래프에 레이블 표시
plt.plot(x, y2, label='수학')
plt.title('class1의 국어와 수학성적')#상단 중앙에 그래프의 제목 표시
plt.legend(loc='upper center') #범례(레이블)을 중앙 상단에 위치시킴 (추가로, 'upper right', 'center left', 'center', 'center right'
                               # 'lower left', 'lower center', 'lower right' , 'best'도 있음)

plt.xlabel('학생 이름') #x축 레이블
plt.ylabel('성적') #y축 레이블
plt.show()
```

```python
#선 스타일 결정
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic') #한글 '맑은 고딕'폰트 사용

x = ['Nana', 'Bo', 'Bibi']
y1 = [65, 80, 100]
y2 = [70, 50, 85]
y3 = [92, 76, 83]
y4 = [90, 86, 80]

#color키워드: red, blue, green, black, white, cyan, magenta, yellow 등 또는 #으로 시작되는 6자리 색상 코드
# 참고로, #은 16진수를 의미

#linestyle 키워드: '-'는 실선, '--'는 파선, '-.'은 일점쇄선, ':'는 점선
plt.plot(x, y1, label='국어', color='red', linestyle='-', marker='o')
plt.plot(x, y2, label='수학', color='cyan', linestyle='--', marker='o')
plt.plot(x, y3, label='영어', color='#00ffff', linestyle='-', marker='o')
plt.plot(x, y4, label='과학', color='#333333', linestyle='-', marker='o')

plt.title('class1 학생들의 성적')
plt.legend(loc='best') #범례를 최적의 위치에

plt.show()
```

```python
#눈금 설정(grid함수 이용)
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic') #한글 '맑은 고딕'폰트 사용

x = ['Nana', 'Bo', 'Bibi']
y1 = [30, 80, 100]
y2 = [70, 50, 85]
y3 = [20, 76, 83]
y4 = [90, 86, 40]

#color키워드: red, blue, green, black, white, cyan, magenta, yellow 등 또는 #으로 시작되는 6자리 색상 코드
# 참고로, #은 16진수를 의미

#linestyle 키워드: '-'는 실선, '--'는 파선, '-.'은 일점쇄선, ':'는 점선
#marker은 o, x, s, d, * 있음
plt.plot(x, y1, label='국어', color='red', linestyle='-', marker='o')
plt.plot(x, y2, label='수학', color='cyan', linestyle='--', marker='x')
plt.plot(x, y3, label='영어', color='#00ffff', linestyle='-', marker='s')
plt.plot(x, y4, label='과학', color='#333333', linestyle='-', marker='d')

plt.title('class1 학생들의 성적')
plt.legend(loc='best') #범례를 최적의 위치에

plt.grid(True) #눈금 설정, 참고로, False면 눈금 사라짐
plt.show()
```
