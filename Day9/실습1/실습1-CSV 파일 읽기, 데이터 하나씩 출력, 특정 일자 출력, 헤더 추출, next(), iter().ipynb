```python
#csv 파일 읽기
import csv
f = open('month_temp.csv','r',encoding='utf8')
lines = csv.reader(f) #python과 다르게 'er'붙임

for line in lines:
    print(line)
f.close()
```

```python
#csv 파일 데이터 하나씩 출력
import csv

f = open('month_temp.csv','r',encoding='utf8')
lines = csv.reader(f) #string으로 받음
for line in lines:
    for x in range(len(line)):
        print(line[x])
f.close()
```

```python
#특정 데이터 출력
import csv
f = open('month_temp.csv','r',encoding='utf8')
lines = csv.reader(f)
for line in lines:
    if '2019-10-14' in line: #10월 14일이 있는 줄 찾기
        print(line[0],line[1][0:7], line[4][0:2]) #지점, 날짜(년과 월만) , 최고온도의 정수값만 출력
f.close()
```

```python
#csv 파일의 헤더 추출(아래 'next() 예시 참고')
import csv
f = open('month_temp.csv','r',encoding='utf8')
lines = csv.reader(f)
header = next(lines)
print(header)
f.close()
```

```python
#next() 함수와 iter()함수 예시
mylist = iter(['빨','주','노','초','파','남','보']) #iter()함수는 반복 가능한 객체를 Iterator객체로 만듦. 
                                                    #이렇게 함으로 next()함수를 이용하여 Iterator 객체의 요소를 하나씩 읽어올 수 있음
x = next(mylist)
print(x) #'빨' 출력
next(mylist)# 저장안해서
print(x) # x는 그대로 '빨' 출력
x=next(mylist)
print(x) #next함수가 3번 돌아서, '노'출력
```
