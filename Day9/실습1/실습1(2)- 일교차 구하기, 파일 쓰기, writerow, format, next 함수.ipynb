```python
#일교차 = 최고기온-최저기온 구하기
import csv 
f = open('month_temp.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

print('%s      %s %s %s'%('일자', '최저기온', '최고기온', '일교차')) #'일자 최저기온 최고기온 일교차' 출력
next(lines) #헤더파일 건너뜀 (많이 씀)
for line in lines:
    diff = float(line[4]) - float(line[3]) 
    print('%s   %.1f   %.1f   %.1f'%(line[1], float(line[3]), float(line[4]),diff))
f.close()
```

```python
#csv 파일 쓰기 (writerow, format, next 함수 이용)
import csv
f1 = open('month_temp.csv', 'r', encoding = 'utf-8')
f2 = open('month_temp2.csv', 'w', encoding = 'utf-8', newline='') #newline=''는 파일을 쓸 때, 행의 끝에 빈 줄을 삽입하지 말라는 옵션

lines = csv.reader(f1)
wr = csv.writer(f2) 

wr.writerow(['일자', '최저기온', '최고기온', '일교차'])
next(lines)#f1의 헤더파일 건너뜀
for line in lines:
    diff = float(line[4]) - float(line[3]) #최고기온에서 최저기온 뺀 일교차
    diff = format(diff,'.1f') #소숫점 첫째자리
    wr.writerow([line[1], float(line[3]), float(line[4]), diff])
    
print('파일 쓰기 완료!')
f1.close()
f2.close()
```
