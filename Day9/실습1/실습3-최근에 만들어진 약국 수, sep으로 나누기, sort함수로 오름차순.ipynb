```python
#서초구의 밝은 약국 찾기
import csv

f = open('pharm_2019.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines) #헤더 저장

for line in lines:
    if line[1]=='서초구' and line[0]=='행복을여는서래약국':
        print(line[0], line[1], line[2], sep='/') # sep='/'이용해서 구분해줌
  
f.close()
```

```python
#서초구 약국 중 2017년 이후에 개설된 약국 수
import csv

f = open('pharm_2019.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines) #헤더 저장

city = '서초구'
total=0 
recent=0
for line in lines:
    if line[1]==city:
        total += 1
        if int(line[3])>20170101:
            recent += 1
            
print('%s의 약국수: %d개' %(city,total))
print('2017년 이후에 개설된 약국 수: %d개' %recent)
```

```python
#추출된 약국 목록 정렬하기 (sort()함수 이용)
import csv

f = open('pharm_2019.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines) #헤더 저장
print('번호', header[0], header[2], sep=',')

data=[]

for line in lines:
    if line[1]=='서초구' and int(line[3]) > 20180101: #서초구이면서, 2018년 이후에 만들어진 약국 목록
        tmp = '%s / %s '%(line[0],line[2])
        data.append(tmp)
        
data.sort() #리스트 data의 요소를 오름차순으로 정렬

number = 1
for x in data:
    print('%d. %s'%(number,x))
    number += 1
f.close()
```
