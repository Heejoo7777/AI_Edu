### DataFrame

```python
# 방법1 : list로 만들기
company1 = [['삼성',2000,'스마트폰'],
            ['현대',1000,'자동차'],
            ['네이버',500,'포털']]
```

```python
df1= pd.DataFrame(company1)
```

```python
df1
```

```python
df1.columns = ['기업명','매출액','대표제품']
```

```python
df1
```

```python
# 방법2: dict로 만들기
company2 = {'기업명':['삼성','현대','네이버'],
            '매출액':[2000,1000,500],
            '대표제품':['스마트폰','자동차','포털']
           }
```

```python
df2 = pd.DataFrame(company2)
```

```python
df2
```

### loc

```python
# 모든 행의 2014년 데이터
df.loc[:,'2014년']
```

```python
# 5~10행의 2014년 데이터
# pandas의 loc는 10이하
df.loc[5:10,'2014년']
```

```python
df
```

```python
# 6~9 구로 금천 노원 도봉
# 기관명, 2014년, 2015년 데이터만 필요
# 데이터 프레임에 저장
# shape (4, 2)
```
