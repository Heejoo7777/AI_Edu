```python
#matplotlib → matlab에 plot을 그리는 library
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits 

digits = load_digits()
sampleNum = 10 # 출력할 샘플 번호를 지정

img = digits.images[sampleNum]
print(img)
```

```python
# 위 값을 그래프로!
plt.figure(figsize=(8,2))
plt.imshow(img, interpolation='nearest', cmap=plt.cm.bone_r) # 빈 곳은 interpolation='nearest'는 옆의 점들의 평균으로
                                                             # cmap은 color map
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title("image {}".format(sampleNum+1))
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
digits = load_digits()

samples = [0, 10, 20, 30, 1, 11, 21, 31]
d = []

for i in range(8):
  d.append(digits.images[samples[i]]) # 파이썬의 list append

plt.figure(figsize=(8,2))

for i in range(8):
  plt.subplot(1, 8, i+1)
  plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
  plt.grid(False); plt.xticks([]); plt.yticks([])
  plt.title("image {}".format(i+1))

  
plt.suptitle("Digit 0&1 image")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
D = np.array(d)
print(D.shape)
```

```python
v = []
for i in range(8):
  v.append(d[i].reshape(64,1)) # d[i].reshape..임!! '.'!! 

plt.figure(figsize=(8,3))

for i in range(8):
  plt.subplot(1, 8, i+1)  # 1행 8열 그리고 index
  plt.imshow(v[i], aspect = 0.4, interpolation='nearest', cmap=plt.cm.bone_r)
  plt.grid(False); plt.xticks([]); plt.yticks([])
  plt.title("vector {}".format(i+1))

plt.suptitle("vectorized image", y=1.05)
plt.tight_layout(w_pad=7)
plt.show()
```

# 데이터 유형 보기- 텐서

```python
from scipy import misc 
img_rgb = misc.face() #  raccon 이미지 한장을 불러옴
img_rgb.shape #데이터의 모양
```

```python
plt.subplot(221) # 2행 2열의 첫번째 이미지
plt.imshow(img_rgb, cmap=plt.cm.gray) # color map 을 gray로
plt.axis("off")
plt.title("RGB color image")

plt.subplot(222) # 2행 2열의 두번째 이미지
plt.imshow(img_rgb[:, :, 0], cmap=plt.cm.Reds) # color map 을 Reds로
plt.axis("off")
plt.title("Red channel")

plt.subplot(223) # 2행 2열의 세번째 이미지
plt.imshow(img_rgb[:, :, 1], cmap=plt.cm.Greens) # color map 을 Greens로
plt.axis("off")
plt.title("Green channel")

plt.subplot(224) # 2행 2열의 네번째 이미지
plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.Blues) # color map 을 Blues로
plt.axis("off")
plt.title("Blue channel")

plt.show()
```
